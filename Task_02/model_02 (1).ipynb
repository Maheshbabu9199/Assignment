```python
# Extracting frames from the video

import cv2
import os
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture("/content/drive/MyDrive/Mahesh_ColabNotebooks/Task_02/Input_Data/test/pexels-artem-podrez-6952256 (Original).mp4")

# Get the frame rate
fps = cap.get(cv2.CAP_PROP_FPS)

# Get the total number of frames
total_frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)

# Create a folder to store the frames
output_dir = os.path.join("/content/drive/MyDrive/Mahesh_ColabNotebooks/Task_02/Output_Data", "frames")
os.makedirs(output_dir, exist_ok=True)

# Iterate over the frames
for i in range(int(total_frames)):

  # Read the frame
  ret, frame = cap.read()

  # Check if the frame is read successfully
  if not ret:
    break

  # Save the frame
  frame_name = os.path.join(output_dir, f"{i}.jpg")
  cv2.imwrite(frame_name, frame)

# Release the capture object
cap.release()

# Close all windows
cv2.destroyAllWindows()
```
